#Question 1: Create a layout with three boxes side by side using Flexbox. The first box should have twice the width of the other two. Provide the HTML and CSS code snippets.

HTML
```
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Flexbox Layout</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>
<div class="container">

    <div class="box box1">Box 1</div>

    <div class="box">Box 2</div>

    <div class="box">Box 3</div>

</div>
</body>
</html>
```
CSS
```
.container{
    display:flex;
    gap:20px;
}

.box{
    flex:1;
    background-color:lightblue;
    padding:40px;
    text-align:center;
    border:1px solid black;
}

.box1{
    flex:2;
}
```
```
Flex Property:
.flex-item {
  flex: <flex-grow> <flex-shrink> <flex-basis>;
}
```
Explanation:

- display: flex; arranges the boxes horizontally.

- gap: 20px; adds space between the boxes.

- flex: 1; gives equal width to Box 2 and Box 3.

- flex: 2; makes the first box twice as wide as the other two.

#Question 2: Using CSS Grid, create a 2×2 grid layout where the first row spans two columns, and the second row has two equal columns. Provide the HTML and CSS code snippets.

CSS Grid is a two-dimensional layout system that allows developers to organize content into rows and columns.

HTML
```
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Grid Layout</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>
<div class="grid-container">

    <div class="header">Header</div>

    <div class="left">Left Box</div>

    <div class="right">Right Box</div>

</div>
</body>
</html>
```
CSS
```
.grid-container{
    display: grid;
    grid-template-columns: 1fr 1fr;
}

.header{
    grid-column: 1 / 3;
    background-color: skyblue;
    padding: 40px;
    text-align: center;
}

.left{
    background-color: lightgreen;
    padding: 40px;
    text-align: center;
}

.right{
    background-color: lightcoral;
    padding: 40px;
    text-align: center;
}
```
Explanation:

- display: grid; enables Grid layout.

- grid-template-columns: 1fr 1fr; creates two equal columns.

- grid-column: 1 / 3; makes the first row span both columns.

- The second row contains two equal-width boxes.

#Question 3: Write a media query that changes the background color of the body to light grey when the screen width is less than 600px.

Media queries allow developers to apply different CSS rules for different screen sizes, making websites responsive.

CSS
```
@media (max-width:600px){

    body{

        background-color:lightgray;

    }

}
```
Explanation:

- max-width:600px means the styles will apply when the screen width is 600 pixels or smaller.

- background-color: lightgray; changes the page background to light grey.

Benefits of Media Queries:

- Makes websites mobile-friendly.

- Improves user experience on different devices.

- Helps create responsive web designs.

- Reduces the need for separate mobile websites.

#Question 4: Create a button that changes its background color to blue and scales up slightly when hovered over, using CSS transitions. Provide the HTML and CSS code snippets.

CSS transitions create smooth animations when the value of a CSS property changes. In this example, the button changes its background color to blue and slightly increases in size when the user hovers the mouse over it.

HTML
```
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Transition Button</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>
  <button class="btn">Hover Me</button>
</body>
</html>
```
CSS
```
.btn{

    background-color: gray;

    color: white;

    padding: 12px 25px;

    border-radius: 5px;

    transition: background-color 0.4s ease, transform 0.4s ease;

}

.btn:hover{

    background-color: blue;

    transform: scale(1.2);

}
```
Explanation:

- transition creates a smooth animation.

- background-color changes from gray to blue.

- transform: scale(1.2) enlarges the button by 20%.

- :hover applies the styles when the mouse pointer is over the button.

#Question 5: Use the ::before pseudo-element to add a decorative icon before each list item in an unordered list with the class name decorated-list. Provide the HTML and CSS code snippets.

The `::before` pseudo-element inserts content before the actual content of an element without modifying the HTML.

#Question 6: Create a simple animation that moves a square box from left to right infinitely. Provide the HTML and CSS code snippets.

CSS animations allow elements to move or change appearance without using JavaScript. In this example, a square box continuously moves from the left side of the screen to the right.

HTML
```
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>CSS Animation</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>
<div class="box"></div>
</body>
</html>
```
CSS
```
.box{
    width:100px;
    height:100px;
    background:red;
    position:relative;

    animation:moveBox 4s linear infinite;
}

@keyframes moveBox{

    from{
        left:0;
        transform:rotate(0deg);
    }

    to{
        left:calc(100vw - 100px);
        transform:rotate(360deg);
    }

}
```
Explanation:

`@keyframes` defines the animation.

`from` specifies the starting position.

`to` specifies the ending position.

`animation: moveBox 4s linear infinite;`

`moveBox` → Animation name.

`4s` → Animation duration.

`linear` → Constant animation speed.

`infinite` → Repeats forever.

Output:

A red square appears.

The square moves from left to right.

The animation repeats continuously without stopping.

#Question 7: Explain the difference between position: absolute; and position: fixed; in CSS.

The position property in CSS controls how an HTML element is placed on a webpage. Two commonly used positioning values are absolute and fixed. Although both remove an element from the normal document flow, they behave differently.

1. position: absolute
An element with position: absolute is positioned relative to its nearest positioned ancestor (an ancestor with position: relative, absolute, or fixed). If no such ancestor exists, it is positioned relative to the webpage.

Example
```
.container{
    position: relative;
}

.box{
    position: absolute;
    top: 20px;
    left: 30px;
}
```
Explanation:

The element is removed from the normal document flow.

It does not reserve space where it originally existed.

It moves relative to its nearest positioned parent.

If no positioned parent is available, it is positioned relative to the entire page.

Common Uses

Dropdown menus

Tooltips

Notification badges

Pop-up elements

2. position: fixed
An element with position: fixed is positioned relative to the browser viewport instead of its parent element. It remains in the same position even when the webpage is scrolled.

Example
```
.chat-button{

    position: fixed;

    bottom: 20px;

    right: 20px;

}
```
Explanation:

The element is removed from the normal document flow.

It always stays visible on the screen.

Scrolling the webpage does not change its position.

Common Uses

Back to Top button

Chat support icon

Sticky contact button

Floating navigation menu


Conclusion:

- Absolute positioning is useful when an element should be positioned inside another element.

- Fixed positioning is useful when an element should always remain visible on the user's screen.

- Choosing the appropriate positioning method helps create better webpage layouts and improves the user experience.

#Question 8: Create a layered layout where a semi-transparent overlay covers an image using z-index. Provide the HTML and CSS code snippets.

The z-index property controls the stacking order of overlapping elements. An element with a higher z-index value appears above an element with a lower value.

In this example, a semi-transparent overlay is placed on top of an image.

HTML
```
<!DOCTYPE html>
<html lang="en">
<head>

    <meta charset="UTF-8">

    <meta name="viewport" content="width=device-width, initial-scale=1.0">

    <title>Z-Index Example</title>

    <link rel="stylesheet" href="style.css">

</head>
<body>
<div class="container">

    <img src="nature.jpg" alt="Nature Image">

    <div class="overlay">

        <h2>Welcome to Nature</h2>

    </div>

</div>
</body>
</html>
```
CSS
```
.container{

    position: relative;

    width: 500px;

}

.container img{

    width: 100%;

    display: block;

}

.overlay{

    position: absolute;

    top: 0;

    left: 0;

    width: 100%;

    height: 100%;

    background-color: rgba(0, 0, 0, 0.5);

    color: white;

    display: flex;

    justify-content: center;

    align-items: center;

    z-index: 1;

}
```
Explanation:
```
1. position: relative
.container{
    position: relative;
}
```
This makes the container the reference point for the overlay.

```
2. position: absolute
.overlay{
    position: absolute;
}
```
This allows the overlay to cover the image completely.

```
3. rgba(0,0,0,0.5)
background-color: rgba(0,0,0,0.5);
```
rgba stands for Red, Green, Blue, Alpha.

The value 0.5 makes the background 50% transparent, allowing the image beneath it to remain visible.

```
4. z-index
z-index:1;
```
The overlay appears above the image because it has a higher stacking order.

```
5. Flexbox
display:flex;
justify-content:center;
align-items:center;
```
These properties center the heading both horizontally and vertically.

Output
The image is displayed normally.

A semi-transparent dark overlay covers the image.

The text appears centered on top of the image.

The overlay remains above the image due to the z-index property.

Conclusion:

The z-index property is used to control the stacking order of elements. Combined with position: relative, position: absolute, and a semi-transparent rgba() background, it allows developers to create modern layered layouts such as image overlays, hero sections, banners, and promotional cards.
