#Question 1 : Explain the basic structure of an HTML page. Why is it important to follow this structure?


```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
</head>
<body>
    <h1>Welcome</h1>
    <p>This is my first HTML page.</p>
</body>
</html>
```

HTML (HyperText Markup Language) is the standard Markup language for a webpage used to define the meaning and structure of web content.
The basic standard structure of html helps to organize and render the web content and disply it on the browser.
Tags with meaning:

1.`<!DOCTYPE html>` - Tells browser that it is a HTML5 document.

2.`<html>` -Root element containing entire webpage.

3.`<head>` - contains information about webpage such as title, style, metadata.

3.`<title>` - to display webpage title in the browser.

4.`<body>` - contains the actual visible webpage content like heading, button, paragraphs, etc.



#Question2 : You’re building a small personal profile page. Which HTML tags will you use to highlight your name as a heading, describe yourself in a paragraph, and emphasize a key skill?

````python
```<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>My Profile</title>
</head>
<body>
  <h1>Gaurav Kumar Mishra</h1>
  <p>Hello there, My name is Gaurav Kumar Mishra. I am from Ranchi, Jharkhand. I am learning web development.</p>
  <p>My key skill is <em>JavaScript</em></p>
</body>
</html>```
````

- `<h1>` tag is used for heading.
- `<p>` tag is used for paragraph.
- `<em>` tag is used to emphasize.

#Question 4: You’re creating a form that asks the user to enter their name, select their country from a list, and click a button to submit. Write the HTML code for this.

````python
```<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Registration Form</title>
</head>
<body>
  <form>

    <label for="name">Name:</label><br>
    <input type="text" name="name" id="name" placeholder="Enter your Name"><br><br>

    <label for="country">Country:</label><br>
    <select name="country" id="country">
        <option value="India">India</option>
        <option value="USA">USA</option>
        <option value="Canada">Canada</option>
        <option value="Australia">Australia</option>
    </select>

    <br><br>

    <button type="submit">Submit</button>

  </form>
</body>
</html>```
````

#Question 5: Write short HTML code to show your favorite fruits in a bulleted list and your top 3 programming languages in a numbered list.

````python

```<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Favourites</title>
</head>
<body>
  <h1>My Favourite Fruits<h1>
  <ul>
    <li>Apple</li>
    <li>Mango</li>
    <li>Pineapple</li>
  </ul>
  <h1>Top 3 programming languages<h1>
    <ol>
      <li>Python</li>
      <li>Javascript</li>
      <li>Java</li>
    </ol>
</body>
</html>```
````
