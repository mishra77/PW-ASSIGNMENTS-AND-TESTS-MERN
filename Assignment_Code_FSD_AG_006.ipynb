#Question 1: Explain what the forEach() method does in JavaScript arrays and write code that uses forEach() to display each student's name from an array of student objects to the console.
Given Array

const students = [
    { name: "Rahul", age: 20 },
    { name: "Priya", age: 19 },
    { name: "Amit", age: 21 },
    { name: "Sneha", age: 20 }
];

## JavaScript `forEach()` Method

The `forEach()` method is an array method that executes a callback function once for each element in an array.

Unlike `map()`, `filter()`, or `reduce()`, the `forEach()` method does **not** return a new array. It is mainly used to perform operations such as displaying data, updating values, or calling another function for every array element.

## JavaScript Code

```javascript
const students = [
    { name: "Rahul", age: 20 },
    { name: "Priya", age: 19 },
    { name: "Amit", age: 21 },
    { name: "Sneha", age: 20 }
];

students.forEach(student => {
    console.log(student.name);
});
```

## Output

```text
Rahul
Priya
Amit
Sneha
```

## Explanation

The statement

```javascript
students.forEach(student => {
    console.log(student.name);
});
```

works as follows:

- `students.forEach()` visits every object in the array.
- During each iteration, the current object is stored in the variable `student`.
- `student.name` accesses the `name` property of the current object.
- `console.log()` prints the student's name.

## Advantages of `forEach()`

- Easy to read and write.
- Eliminates the need for manual indexing.
- Useful for displaying or updating data.
- Executes the callback function once for every array element.

## Conclusion

The `forEach()` method is best used when performing an operation on every element of an array without creating a new array.

#Question 2: You have an array of product prices in rupees. Write code using the map() method to create a new array that contains the same prices but with 18% GST added to each price.

##JavaScript `map()` Method

The `map()` method creates a new array by applying a function to every element of an existing array.

## JavaScript Code

```javascript
const originalPrices = [100, 250, 500, 750, 1000];

const pricesWithGST = originalPrices.map(price => {

    return price + (price * 18 / 100);

});

console.log(pricesWithGST);
```

## Explanation

The statement

```javascript
originalPrices.map(price => {

    return price + (price * 18 / 100);

});
```

works as follows:

- `map()` processes each element (`price`) in the array.
- Calculates **18% GST** for each price.
- Adds the GST to the original price.
- Returns the updated price.
- Stores all updated prices in a **new array**.


## Advantages of `map()`

- Does not modify the original array.
- Returns a new array.
- Makes data transformation simple and readable.

## Conclusion

Use the `map()` method whenever every element of an array needs to be transformed into a new value while keeping the original array unchanged.

#Question 3: A school wants to find all students who scored more than 80 marks. The following code has an error that prevents it from working correctly. Fix the code and explain what was wrong.
Given Code

const students = [
    { name: "Raj", marks: 85 },
    { name: "Sita", marks: 92 },
    { name: "Vikram", marks: 78 },
    { name: "Meera", marks: 88 }
];

const topStudents = students.filter(student => {
    student.marks > 80;
});

## JavaScript `filter()` Method

The error is that the callback function inside `filter()` does not return the condition.

## Correct Code

```javascript
const students = [
    { name: "Raj", marks: 85 },
    { name: "Sita", marks: 92 },
    { name: "Vikram", marks: 78 },
    { name: "Meera", marks: 88 }
];

const topStudents = students.filter(student => {

    return student.marks > 80;

});

console.log(topStudents);
```

## Alternative (Short Arrow Function)

```javascript
const topStudents = students.filter(student => student.marks > 80);
```

## Output

```text
[
  { name: "Raj", marks: 85 },
  { name: "Sita", marks: 92 },
  { name: "Meera", marks: 88 }
]
```

## What Was Wrong?

The original code was:

```javascript
students.filter(student => {

    student.marks > 80;

});
```

Since curly braces `{}` are used, JavaScript expects an explicit `return` statement.

Without `return`, the callback returns `undefined` for every element.

As a result, `filter()` treats every condition as `false`, and the resulting array becomes empty.

## How `filter()` Works

The `filter()` method:

- Checks every element in the array.
- Returns only those elements whose condition evaluates to `true`.
- Creates a new filtered array.
- Does not modify the original array.

## Why Use `filter()`?

- Selects only the required elements.
- Produces clean and readable code.
- Commonly used in searching, filtering, and validation operations.

## Conclusion

The `filter()` method requires the callback function to return a Boolean value. Adding the missing `return` statement (or using the concise arrow function syntax) fixes the error and correctly returns the students who scored more than `80` marks.

## JavaScript `find()` Method

The `find()` method searches an array and returns the **first element** that satisfies the given condition. If no matching element is found, it returns `undefined`.

## JavaScript Code

```javascript
const books = [
    { title: "Gitanjali", author: "Rabindranath Tagore" },
    { title: "Harry Potter", author: "J.K. Rowling" },
    { title: "The God of Small Things", author: "Arundhati Roy" },
    { title: "Midnight's Children", author: "Salman Rushdie" }
];

const book = books.find(item => item.title === "Harry Potter");

if (book) {
    console.log(book.author);
} else {
    console.log("Book not found");
}
```

## Output

```text
J.K. Rowling
```

## Explanation

```javascript
books.find(item => item.title === "Harry Potter");
```

- `find()` checks each object in the array.
- It compares the `title` property with `"Harry Potter"`.
- As soon as a matching object is found, it returns that object.

Then,

```javascript
console.log(book.author);
```

prints the author's name.

If no matching book is found, the `else` block executes and displays:

```text
Book not found
```

## Advantages of `find()`

- Returns only the **first** matching element.
- Stops searching once a match is found.
- Makes searching arrays simple and efficient.

## Conclusion

The `find()` method is useful when you need to locate a single element in an array based on a specific condition.

## JavaScript `sort()` Method

## JavaScript Code

```javascript
const employees = [
    { name: "Rajesh", salary: 45000 },
    { name: "Priya", salary: 55000 },
    { name: "Amit", salary: 40000 },
    { name: "Kavya", salary: 60000 }
];

employees.sort((a, b) => b.salary - a.salary);

console.log(employees);
```

## Explanation

```javascript
employees.sort((a, b) => b.salary - a.salary);
```

- `a` and `b` represent two employee objects.
- `b.salary - a.salary` sorts the salaries in **descending order** (highest to lowest).
- The original array is modified after sorting.


## Advantages of `sort()`

- Sorts arrays alphabetically or numerically.
- Can sort objects using custom comparison functions.
- Useful for ranking, leaderboards, and reports.

## Conclusion

The `sort()` method organizes data efficiently. Using a comparison function allows accurate sorting of numeric values and object properties.

#Question 6: Write code using the reduce() method to calculate the total value of all items in a shopping cart. Each item has a price and quantity. Also find the most expensive single item in the cart.
Given Array

const cart = [
    { name: "Laptop", price: 50000, quantity: 1 },
    { name: "Mouse", price: 500, quantity: 2 },
    { name: "Keyboard", price: 1500, quantity: 1 },
    { name: "Monitor", price: 15000, quantity: 1 }
];

# JavaScript `reduce()` Method

The `reduce()` method processes all elements of an array and reduces them to a single value, such as a total, average, or maximum value.

## JavaScript Code

```javascript
const cart = [
    { name: "Laptop", price: 50000, quantity: 1 },
    { name: "Mouse", price: 500, quantity: 2 },
    { name: "Keyboard", price: 1500, quantity: 1 },
    { name: "Monitor", price: 15000, quantity: 1 }
];

// Calculate total cart value
const totalValue = cart.reduce((total, item) => {
    return total + (item.price * item.quantity);
}, 0);

console.log("Total Value:", totalValue);

// Find the most expensive item
const expensiveItem = cart.reduce((highest, item) => {
    return item.price > highest.price ? item : highest;
});

console.log("Most Expensive Item:", expensiveItem);
```

## Output

```text
Total Value: 67500

Most Expensive Item:
{
  name: "Laptop",
  price: 50000,
  quantity: 1
}
```

## Explanation

### Calculating Total Value

```javascript
cart.reduce((total, item) => {
    return total + (item.price * item.quantity);
}, 0);
```



---

### Finding the Most Expensive Item

```javascript
cart.reduce((highest, item) => {
    return item.price > highest.price ? item : highest;
});
```

- Compares the price of each item.
- Returns the object with the highest price.
- In this case, **Laptop** is the most expensive item.

## Advantages of `reduce()`

- Calculates totals, averages, and maximum values.
- Processes arrays into a single result.
- Reduces the need for manual loops.
- Makes code concise and readable.

## Conclusion

The `reduce()` method is one of the most powerful array methods in JavaScript. It can perform calculations, accumulate values, and even find the maximum or minimum element in an array efficiently.

## JavaScript Array Destructuring

Array destructuring is an **ES6** feature that allows us to extract values from an array and assign them to separate variables in a simple and readable way.

Instead of accessing elements using indexes like `colors[0]` or `colors[1]`, destructuring lets us assign multiple values in a single statement.

## JavaScript Code

```javascript
const colors = ["red", "blue", "green", "yellow", "purple"];

// Array Destructuring
const [firstColor, secondColor] = colors;

console.log("First Color:", firstColor);
console.log("Second Color:", secondColor);
```

## Output

```text
First Color: red
Second Color: blue
```

## Explanation

The statement

```javascript
const [firstColor, secondColor] = colors;
```


## Advantages of Array Destructuring

- Makes code shorter and cleaner.
- Improves readability.
- Allows extraction of multiple values at once.
- Frequently used with arrays returned by functions.

## Conclusion

Array destructuring provides a simple and efficient way to access array elements without using index numbers.

#Question 9: Write a function called calculateTotal that can accept any number of prices as arguments and returns their sum. Use the rest operator to handle the variable number of parameters.

## JavaScript Rest Operator (`...`)

The **rest operator (`...`)** is an ES6 feature that allows a function to accept any number of arguments. These arguments are collected into a single array, making it easy to process multiple values.

## JavaScript Code

```javascript
function calculateTotal(...prices) {

    return prices.reduce((total, price) => total + price, 0);

}

console.log(calculateTotal(100, 200, 300));

console.log(calculateTotal(50, 150, 250, 350));

console.log(calculateTotal(500));
```

## Output

```text
600
800
500
```

## Explanation

### Function Declaration

```javascript
function calculateTotal(...prices)
```

- `...prices` is the **rest parameter**.
- It collects all function arguments into an array named `prices`.

For example,

```javascript
calculateTotal(100, 200, 300);
```

becomes

```javascript
prices = [100, 200, 300];
```

### Using `reduce()`

```javascript
prices.reduce((total, price) => total + price, 0);
```

- Starts with an initial value of `0`.
- Adds each price to the running total.
- Returns the final sum.

## Advantages of the Rest Operator

- Accepts any number of arguments.
- Makes functions more flexible.
- Eliminates the need to know the number of parameters beforehand.
- Produces cleaner and more readable code.

## Conclusion

The rest operator is useful when a function needs to handle a variable number of arguments. Combined with `reduce()`, it provides an efficient way to calculate totals.
