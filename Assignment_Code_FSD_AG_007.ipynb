#Question 1: Write code that creates a new paragraph element with the text "Welcome to our website!" and adds it to a div with the id "content" on the webpage. Explain what createElement() and appendChild() methods do.

## JavaScript DOM: `createElement()` and `appendChild()`

The **Document Object Model (DOM)** allows JavaScript to create, modify, and remove HTML elements dynamically. In this example, a new paragraph element is created and added to a `<div>` with the id `content`.

## HTML

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>DOM Example</title>
</head>

<body>

    <div id="content"></div>

    <script src="script.js"></script>

</body>
</html>
```

## JavaScript

```javascript
// Find the div element
const content = document.getElementById("content");

// Create a new paragraph element
const paragraph = document.createElement("p");

// Add text to the paragraph
paragraph.textContent = "Welcome to our website!";

// Append the paragraph to the div
content.appendChild(paragraph);
```

## Output

```text
Welcome to our website!
```

*(A new paragraph appears inside the `<div>`.)*

## Explanation

### 1. `document.createElement()`

```javascript
const paragraph = document.createElement("p");
```

Creates a new HTML element.

The element exists only in memory until it is added to the webpage.

### 2. `textContent`

```javascript
paragraph.textContent = "Welcome to our website!";
```

Adds text inside the paragraph.

### 3. `appendChild()`

```javascript
content.appendChild(paragraph);
```

Adds the newly created paragraph as the last child of the `<div>`.

## Advantages

- Creates HTML elements dynamically.
- Makes webpages interactive.
- Reduces the need to write HTML manually.

## Conclusion

`createElement()` creates new HTML elements, while `appendChild()` inserts them into the webpage. Together, they allow developers to modify web pages dynamically using JavaScript.

#Question 5: Write a function called repeatMessage that takes a message string and a number, and returns the message repeated that many times, separated by a space. Do not use loops.

## JavaScript `repeat()` Method

The `repeat()` method repeats a string a specified number of times. By appending a space after the message and removing the trailing space using `trim()`, the required output can be generated without using loops.

## JavaScript Code

```javascript
function repeatMessage(message, number) {
    return (message + " ").repeat(number).trim();
}

console.log(repeatMessage("hello", 3));
```

## Output

```text
hello hello hello
```

## Explanation

```javascript
(message + " ")
```

Appends a space after the message.

```javascript
.repeat(number)
```

Repeats the string the specified number of times.

```javascript
.trim()
```

Removes the extra space from the beginning and end of the resulting string. In this example, it removes the trailing space.

## Execution

```text
(message + " ")
→ "hello "

.repeat(3)
→ "hello hello hello "

.trim()
→ "hello hello hello"
```

## Conclusion

The `repeat()` and `trim()` string methods provide a simple and efficient way to repeat a message multiple times without using loops.

## JavaScript Event Loop

The **Event Loop** is a mechanism in JavaScript that enables asynchronous operations to execute without blocking the main thread. It continuously monitors the **Call Stack** and the **Callback Queue**. When the Call Stack becomes empty, the Event Loop moves pending callback functions from the Callback Queue to the Call Stack for execution.

## JavaScript Code

```javascript
console.log("Start");

setTimeout(() => {
    console.log("Timeout 1");
}, 0);

console.log("Middle");

setTimeout(() => {
    console.log("Timeout 2");
}, 0);

console.log("End");
```

## Output

```text
Start
Middle
End
Timeout 1
Timeout 2
```

## Explanation

```javascript
console.log("Start");
```

Executes immediately.

```javascript
setTimeout(() => {
    console.log("Timeout 1");
}, 0);
```

Registers its callback with the **Web API**. Even with a delay of `0` milliseconds, the callback is placed in the **Callback Queue** only after the timer expires.

```javascript
console.log("Middle");
```

Executes immediately.

```javascript
setTimeout(() => {
    console.log("Timeout 2");
}, 0);
```

Registers the second callback with the **Web API**.

```javascript
console.log("End");
```

Executes immediately.

After the **Call Stack** becomes empty, the **Event Loop** moves the first callback from the **Callback Queue** to the **Call Stack**, printing:

```text
Timeout 1
```

Finally, the second callback is executed, printing:

```text
Timeout 2
```

## Conclusion

Although the delay is `0` milliseconds, `setTimeout()` callbacks execute only after all synchronous code has finished because they must wait in the **Callback Queue** until the **Call Stack** is empty.

#Question 7: Write a JavaScript function named startCountdown that logs "Time left: X" every second, starting from 5 down to 1, and finally logs "Time's up!". Use setInterval and clearInterval only.

# JavaScript `setInterval()` and `clearInterval()`

The `setInterval()` function repeatedly executes a block of code at a specified time interval, while `clearInterval()` stops the repeated execution.

## JavaScript Code

```javascript
function startCountdown() {
    let count = 5;

    const timer = setInterval(function () {
        console.log("Time left: " + count);
        count--;

        if (count === 0) {
            clearInterval(timer);
            console.log("Time's up!");
        }
    }, 1000);
}

startCountdown();
```

## Output

```text
Time left: 5
Time left: 4
Time left: 3
Time left: 2
Time left: 1
Time's up!
```

## Explanation

```javascript
setInterval(function () {
    ...
}, 1000);
```

Executes the callback function every **1000 milliseconds (1 second)**.

```javascript
console.log("Time left: " + count);
count--;
```

Displays the current value of `count` and then decrements it after each execution.

```javascript
if (count === 0) {
    clearInterval(timer);
    console.log("Time's up!");
}
```

When `count` becomes `0`, `clearInterval(timer)` stops the interval, and `"Time's up!"` is displayed.

## Conclusion

The `setInterval()` function is useful for performing tasks repeatedly at fixed intervals, while `clearInterval()` is used to stop the execution when it is no longer required.

#Question 8: Explain what Promises are in JavaScript and write code that creates a Promise to simulate checking if a user is logged in. The Promise should resolve with user data after 2 seconds if successful, or reject with an error message if the user is not found.

# JavaScript Promises

A **Promise** is an object in JavaScript that represents the eventual completion or failure of an asynchronous operation. A Promise has three states:

- **Pending** – Initial state.
- **Fulfilled (Resolved)** – The operation completed successfully.
- **Rejected** – The operation failed.

## JavaScript Code

```javascript
function checkUser(isLoggedIn) {
    return new Promise((resolve, reject) => {
        setTimeout(() => {
            if (isLoggedIn) {
                resolve({
                    name: "Rahul",
                    age: 21
                });
            } else {
                reject("User not found");
            }
        }, 2000);
    });
}

checkUser(true)
    .then(user => {
        console.log(user);
    })
    .catch(error => {
        console.log(error);
    });
```

## Output (If User Exists)

```text
{
  name: "Rahul",
  age: 21
}
```

## Output (If User Does Not Exist)

```text
User not found
```

## Explanation

```javascript
new Promise((resolve, reject) => { ... });
```

Creates a new Promise object.

```javascript
resolve({
    name: "Rahul",
    age: 21
});
```

Returns the user data after a delay of **2 seconds** when the operation is successful.

```javascript
reject("User not found");
```

Returns an error message if the operation fails.

```javascript
.then(user => {
    console.log(user);
});
```

Handles the **fulfilled (resolved)** state of the Promise.

```javascript
.catch(error => {
    console.log(error);
});
```

Handles the **rejected** state and processes any errors.

## Conclusion

Promises simplify asynchronous programming by providing a structured way to handle successful results with `.then()` and errors with `.catch()`, making code more readable and maintainable.

Given Code

function getUserDataAndPosts(userId) {
    return fetchUserData(userId)
        .then(userData => {

# JavaScript `async` and `await`

The `async`/`await` syntax provides a cleaner and more readable way to write asynchronous code. It allows asynchronous operations to be written in a style similar to synchronous code while still working with Promises.

## JavaScript Code

```javascript
async function getUserDataAndPosts(userId) {
    try {
        const userData = await fetchUserData(userId);
        const posts = await fetchUserPosts(userData.id);

        return {
            user: userData,
            posts: posts
        };
    } catch (error) {
        console.error("Error:", error);
        throw error;
    }
}
```

## Explanation

```javascript
async function getUserDataAndPosts(userId)
```

The `async` keyword declares an asynchronous function, which automatically returns a Promise.

```javascript
const userData = await fetchUserData(userId);
```

The `await` keyword pauses the execution of the function until the Promise returned by `fetchUserData()` is resolved.

```javascript
const posts = await fetchUserPosts(userData.id);
```

Waits for the user's posts to be retrieved before continuing.

```javascript
try {
    ...
} catch (error) {
    ...
}
```

The `try...catch` block handles any errors that occur during the asynchronous operations.

```javascript
return {
    user: userData,
    posts: posts
};
```

Returns an object containing both the user data and the user's posts.

## Conclusion

Using `async`/`await` makes asynchronous code easier to read, understand, and maintain compared to chaining multiple `.then()` methods.

#Question 10: Write code using the Fetch API to get weather data from a public API. Display the weather information in a user-friendly format. Assume you have the following configuration for the API.

const apiKey = "abcd1234";

# JavaScript Fetch API

The **Fetch API** is used to send HTTP requests and receive responses from a server. In this example, it retrieves weather data from the OpenWeatherMap API.

## JavaScript Code

```javascript
const apiKey = "abcd1234";

const apiUrl =
`https://api.openweathermap.org/data/2.5/weather?q=Mumbai&appid=${apiKey}&units=metric`;

fetch(apiUrl)
    .then(response => response.json())
    .then(data => {
        console.log("City:", data.name);
        console.log("Temperature:", data.main.temp + "°C");
        console.log("Weather:", data.weather[0].description);
        console.log("Humidity:", data.main.humidity + "%");
    })
    .catch(error => {
        console.log("Error:", error);
    });
```

## Sample Output

```text
City: Mumbai
Temperature: 31°C
Weather: clear sky
Humidity: 74%
```

## Explanation

```javascript
fetch(apiUrl)
```

Sends an HTTP request to the weather API.

```javascript
response.json()
```

Parses the JSON response and returns a Promise that resolves to a JavaScript object.

```javascript
.then(data => {
    ...
});
```

Processes the response data and displays the weather details using `console.log()`.

```javascript
.catch(error => {
    console.log("Error:", error);
});
```

Handles errors such as network failures or invalid API requests.

## Conclusion

The Fetch API provides a simple and efficient way to retrieve data from web services. It is widely used to communicate with REST APIs and display dynamic data on web pages.
