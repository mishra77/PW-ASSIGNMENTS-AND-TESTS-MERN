#Question 3: Write a function getDayName that takes a number (1–7) and returns the corresponding day of the week. Use a switch statement.

## JavaScript `switch` Statement

A `switch` statement is used when there are multiple possible values for a single expression. It provides a cleaner alternative to writing multiple `if-else` statements.

## JavaScript Code

```javascript
function getDayName(day) {

    switch (day) {

        case 1:
            return "Monday";

        case 2:
            return "Tuesday";

        case 3:
            return "Wednesday";

        case 4:
            return "Thursday";

        case 5:
            return "Friday";

        case 6:
            return "Saturday";

        case 7:
            return "Sunday";

        default:
            return "Invalid Day";

    }

}

console.log(getDayName(3));

console.log(getDayName(7));

console.log(getDayName(10));
```

## Output

```text
Wednesday
Sunday
Invalid Day
```

## Explanation

### Switch Statement

```javascript
switch(day)
```

The value of `day` is compared with each `case`.

### Case Statements

Each `case` represents a day of the week.

Example:

```javascript
case 1:
    return "Monday";
```

If `day` equals `1`, the function returns `"Monday"`.

### Default Case

```javascript
default:
    return "Invalid Day";
```

The `default` block executes when none of the cases match.

For example:

```javascript
getDayName(10);
```

returns

```text
Invalid Day
```

## Advantages of Using `switch`

- Easier to read than multiple `if-else` statements.
- Ideal when checking one variable against many possible values.
- Improves code organization.

## Conclusion

The `switch` statement is an efficient way to select one action from multiple possible options based on the value of a variable.

## JavaScript Ternary Operator (`? :`)

The **ternary operator (`? :`)** is a shorthand way of writing an `if-else` statement. It evaluates a condition and returns one value if the condition is true and another value if it is false.

## JavaScript Code

```javascript
function checkEligibility(age) {

    return (age >= 18) ? "Eligible" : "Not Eligible";

}

console.log(checkEligibility(20));

console.log(checkEligibility(15));
```

## Output

```text
Eligible
Not Eligible
```

## Explanation

### Ternary Operator Syntax

```javascript
condition ? expression_if_true : expression_if_false;
```

In this example:

```javascript
(age >= 18) ? "Eligible" : "Not Eligible";
```

- If `age >= 18` is true, it returns `"Eligible"`.
- Otherwise, it returns `"Not Eligible"`.

## Advantages

- Shorter than an `if-else` statement.
- Improves readability for simple conditions.
- Commonly used in modern JavaScript.

## Conclusion

The ternary operator is useful for simple decision-making and makes the code concise and easy to understand.

#Question 5: Given an array numbers = [1, 2, 3, 4, 5], perform the following:
Use a for loop to print each number.

Use a while loop to print each number.

Use a for-in loop to print each index.

Use a for-of loop to print each number.

## JavaScript Loops

JavaScript provides different types of loops to iterate through arrays and other collections.

## JavaScript Code

```javascript
const numbers = [1, 2, 3, 4, 5];

// 1. for loop
console.log("Using for loop:");

for (let i = 0; i < numbers.length; i++) {

    console.log(numbers[i]);

}

// 2. while loop
console.log("Using while loop:");

let i = 0;

while (i < numbers.length) {

    console.log(numbers[i]);

    i++;

}

// 3. for-in loop
console.log("Using for-in loop:");

for (let index in numbers) {

    console.log(index);

}

// 4. for-of loop
console.log("Using for-of loop:");

for (let value of numbers) {

    console.log(value);

}
```

## Output

```text
Using for loop:
1
2
3
4
5

Using while loop:
1
2
3
4
5

Using for-in loop:
0
1
2
3
4

Using for-of loop:
1
2
3
4
5
```

## Explanation

### 1. `for` Loop

```javascript
for (let i = 0; i < numbers.length; i++)
```

Uses an index variable.

Best when you need the position (index) of elements.

### 2. `while` Loop

```javascript
while (i < numbers.length)
```

Continues executing until the condition becomes false.

Useful when the number of iterations is not fixed.

### 3. `for...in` Loop

```javascript
for (let index in numbers)
```

Returns the **indexes** of the array.

**Output:**

```text
0
1
2
3
4
```

### 4. `for...of` Loop

```javascript
for (let value of numbers)
```

Returns the **actual values** of the array.

Simpler and cleaner for reading array elements.

## Conclusion

Different loops serve different purposes:

- **`for`** → Index-based iteration.
- **`while`** → Condition-based iteration.
- **`for...in`** → Iterates over indexes.
- **`for...of`** → Iterates over values.

## JavaScript Arrow Functions

An arrow function is a shorter and more modern way of writing functions in JavaScript. It was introduced in **ES6 (ECMAScript 2015)** and is commonly used because of its simple syntax.

## Traditional Function

```javascript
function add(num1, num2) {

    return num1 + num2;

}
```

## Arrow Function

```javascript
const add = (num1, num2) => {

    return num1 + num2;

};

console.log(add(10, 20));
```

## Output

```text
30
```

## Explanation

### Arrow Function Syntax

```javascript
const functionName = (parameters) => {

    // code

};
```

In this example:

```javascript
const add = (num1, num2) => {

    return num1 + num2;

};
```

- `const add` creates a function named `add`.
- `(num1, num2)` are the function parameters.
- `=>` is the arrow function operator.
- `return` returns the sum of the two numbers.

## Advantages of Arrow Functions

- Shorter syntax.
- Easier to read.
- Frequently used in callbacks and array methods.
- Automatically inherit `this` from the surrounding scope.

## Conclusion

Arrow functions provide a cleaner and more concise way of writing functions, making JavaScript code easier to read and maintain.

#Question 8: Write a function greet that takes a name and a callback function. The greet function should call the callback with the name.

## JavaScript Callback Functions

A callback function is a function passed as an argument to another function. It allows one function to execute another function after completing a task.

## JavaScript Code

```javascript
function greet(name, callback) {

    callback(name);

}

function displayMessage(name) {

    console.log("Hello, " + name + "!");

}

greet("Rahul", displayMessage);
```

## Output

```text
Hello, Rahul!
```

## Explanation

### `greet` Function

```javascript
function greet(name, callback)
```

The function accepts two parameters:

- `name`
- `callback`

### Calling the Callback

```javascript
callback(name);
```

The callback function is executed, passing the value of `name`.

### Callback Function

```javascript
function displayMessage(name) {

    console.log("Hello, " + name + "!");

}
```

This function receives the `name` and prints a greeting message.

### Function Call

```javascript
greet("Rahul", displayMessage);
```

**Execution Flow:**

1. `greet()` is called.
2. `"Rahul"` is passed as the `name`.
3. `displayMessage()` is passed as the callback.
4. `callback(name)` executes `displayMessage("Rahul")`.
5. The greeting message is displayed.

## Advantages of Callback Functions

- Promote code reusability.
- Make programs more flexible.
- Used extensively in asynchronous programming.
- Commonly used with timers, event handlers, and API requests.

## Conclusion

Callback functions enable one function to execute another function, making JavaScript programs more modular and efficient.

## JavaScript Array `length` Property

Arrays in JavaScript have a built-in **`length`** property that returns the total number of elements in an array. It can also be used to add new elements at the end of the array.

## JavaScript Code

```javascript
const fruits = ["apple", "banana", "cherry"];

// Print the length
console.log(fruits.length);

// Add a new element using the length property
fruits[fruits.length] = "date";

// Print the updated array
console.log(fruits);
```

## Output

```text
3

["apple", "banana", "cherry", "date"]
```

## Explanation

### Array Length

```javascript
fruits.length
```

Returns the total number of elements in the array.

Initially, the array contains:

```text
apple
banana
cherry
```

So, the length is:

```text
3
```

### Adding an Element

```javascript
fruits[fruits.length] = "date";
```

Since:

```text
fruits.length = 3
```

the new element is added at **index 3**, which is the next available position.

The updated array becomes:

| Index | Element |
|:-----:|---------|
| 0 | apple |
| 1 | banana |
| 2 | cherry |
| 3 | date |

## Why Use the `length` Property?

- Returns the total number of elements in an array.
- Automatically adds new elements to the end of the array.
- Eliminates the need to manually calculate the last index.
- Useful when dynamically adding elements.

## Conclusion

The **`length`** property is useful for determining the size of an array and for dynamically adding new elements to the end of the array.

## Output

## Explanation

### 1. `trim()`

```javascript
text.trim();
```

The `trim()` method removes extra spaces from the beginning and end of a string.

Before trimming:

```text
" Hello World! "
```

After trimming:

```text
"Hello World!"
```

---

### 2. `toUpperCase()`

```javascript
trimmedText.toUpperCase();
```

Converts all characters in the string to uppercase.

**Output:**

```text
HELLO WORLD!
```

---

### 3. `includes()`

```javascript
trimmedText.includes("World");
```

Checks whether the string contains `"World"`.

If found, it returns:

```text
true
```

Otherwise, it returns:

```text
false
```

## Why Are These Methods Useful?

- `trim()` cleans user input by removing unnecessary spaces.
- `toUpperCase()` converts text to uppercase for consistent comparisons.
- `includes()` checks whether a string contains a specific word or character.

## Conclusion

JavaScript string methods make it easy to clean, modify, and search text, improving data handling and user input validation.

## JavaScript Array Methods

JavaScript arrays provide several built-in methods to add, remove, search, and test elements efficiently.

## JavaScript Code

```javascript
const numbers = [1, 2, 3, 4, 5];

// 1. Add 6 to the end
numbers.push(6);
console.log(numbers);

// 2. Remove the first element
numbers.shift();
console.log(numbers);

// 3. Find the index of 4
console.log(numbers.indexOf(4));

// 4. Check if all numbers are greater than 0
console.log(numbers.every(num => num > 0));
```

## Output

```text
[1, 2, 3, 4, 5, 6]

[2, 3, 4, 5, 6]

2

true
```

## Explanation

### 1. `push()`

```javascript
numbers.push(6);
```

Adds a new element to the end of the array.

Before:

```text
[1, 2, 3, 4, 5]
```

After:

```text
[1, 2, 3, 4, 5, 6]
```

---

### 2. `shift()`

```javascript
numbers.shift();
```

Removes the first element of the array.

After removal:

```text
[2, 3, 4, 5, 6]
```

---

### 3. `indexOf()`

```javascript
numbers.indexOf(4);
```

Returns the index of the specified element.



Therefore,

```javascript
numbers.indexOf(4);
```

returns:

```text
2
```

---

### 4. `every()`

```javascript
numbers.every(num => num > 0);
```

Checks whether every element satisfies the given condition.

Since all elements are greater than `0`, the result is:

```text
true
```

## Why These Methods Are Useful

- `push()` → Adds elements to the end of an array.
- `shift()` → Removes the first element.
- `indexOf()` → Finds the position (index) of an element.
- `every()` → Tests whether all elements satisfy a given condition.

Array methods simplify common operations such as adding, removing, searching, and validating data. They are widely used in JavaScript applications to manipulate collections of data efficiently.
